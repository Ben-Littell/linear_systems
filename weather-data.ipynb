```python
import matplotlib.pyplot as plt
import numpy as np

file_name = '../input/portland-weather/PortlandWeather2013b.txt'
def open_file(filename):
    new_dict = {}
    key_list = []
    with open(filename) as file:
        text = file.readlines()
    for key in text[0].split()[1:]:
        new_dict.update({key: []})
        key_list.append(key)
    for line in text[2:]:
        counter = 0
        for val in line.split()[1:]:
            if counter == 0:
                new_dict[key_list[counter]].append(val)
            elif counter == 4 or counter == 5:
                new_numb = c_to_f(int(val)/10)
                new_dict[key_list[counter]].append(new_numb)
            else:
                new_dict[key_list[counter]].append(int(val)/10)
            counter += 1
        
    return new_dict
    
def c_to_f(data):
    conv = (data*9/5)+32
    return conv    

def graph_scatter_2sub(x1, y1, title1, x_label1, y_label1, x2, y2, title2, x_label2, y_label2):
    fig = plt.figure(figsize = (8, 8))
    ax1, ax2 = fig.subplots(nrows=2, ncols=1, sharex = True, sharey=True)
    ax1.scatter(x1, y1)
    ax2.scatter(x1, y2)
    ax1.set_title(title1)
    ax2.set_xlabel(x_label1)
    ax1.set_ylabel(y_label1)
    ax2.set_ylabel(y_label2)
    ax2.set_title(title2)
    fig.tight_layout()  
    plt.show()
    
def graph_plot(x1, y1, title1, x_label1, y_label1, x2, y2, label1, label2):
#     fig = plt.figure(figsize = (8, 8))
#     ax1, ax2 = fig.subplots(nrows=2, ncols=1, sharex = True, sharey=True)
    plt.plot(x1, y1, 'g', label=label1)
    plt.plot(x1, y2, 'r', label=label2)
    plt.title(title1)
    plt.xlabel(x_label1)
    plt.ylabel(y_label1)
    plt.legend()
    plt.tight_layout()  
    plt.show()
    
    
def moving_avg(data, interval):
    new_list = []
    for i in range(interval, len(data)):
        k = []
        for j in range(interval+1):
            k.append(data[i-j])
        new_list.append(round(np.average(k), 1))
    return new_list


weather_dict = open_file(file_name)
x_ticks = np.arange(1, 366)


graph_scatter_2sub(x_ticks, weather_dict['TMIN'], 'MIN TEMPS', 'DATE', 'TEMP(F)', x_ticks, weather_dict['TMAX'], 'MAX TEMPS', 'DATE', 'TEMP(F)')
graph_plot(x_ticks, weather_dict['TMIN'], 'TEMPS', 'DATE', 'TEMP(F)', x_ticks, weather_dict['TMAX'], 'MIN', 'MAX',)

mov_avg_min = moving_avg(weather_dict['TMIN'], 10)
mov_avg_max = moving_avg(weather_dict['TMAX'], 10)
x_ticks_mav = np.arange(1, len(mov_avg_min)+1)

graph_plot(x_ticks_mav, mov_avg_min, 'MOVING AVG', 'DATE', 'TEMP(F)', x_ticks_mav, mov_avg_max, 'MIN', 'MAX')
```
